```python
from sklearn.feature_extraction.text import CountVectorizer
from sklearn.feature_extraction.text import TfidfVectorizer

from sklearn.tree import DecisionTreeClassifier

from sklearn.model_selection import train_test_split
from sklearn.tree import DecisionTreeClassifier


import pandas as pd
import numpy as np

import prepare
```

```python
df = pd.read_csv('spam_clean.csv')
```

```python
df.head()
```

```python
#basic clean
df['tidy_text'] = df.text.apply(prepare.basic_clean)
```

```python
#tokenize
df['tidy_text'] = df.tidy_text.apply(prepare.tokenize_string)
```

```python
#remove english stop words
df['tidy_text'] = df.tidy_text.apply(prepare.adjust_stopwords)
```

```python
df.head()
```

```python
cv = CountVectorizer()
tdif = TfidfVectorizer()
```

```python
X = cv.fit_transform(df.tidy_text)
X2 = tdif.fit_transform(df.tidy_text)
y = df.label
```

```python
pd.DataFrame(X.todense(), columns=cv.get_feature_names())
```

```python
pd.DataFrame(X2.todense(), columns=tdif.get_feature_names())
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=.2, random_state=12)

tree = DecisionTreeClassifier(max_depth=5)
tree.fit(X_train, y_train)

tree.score(X_train, y_train)
```

```python
X_train, X_test, y_train, y_test = train_test_split(X2, y, test_size=.2, random_state=12)

tree = DecisionTreeClassifier(max_depth=5)
tree.fit(X_train, y_train)

tree.score(X_train, y_train)
```

```python
#1 and 2 word combo
cv = CountVectorizer(ngram_range=(2,2))
tdif = TfidfVectorizer(ngram_range=(2,2))
```

```python
X = cv.fit_transform(df.tidy_text)
X2 = tdif.fit_transform(df.tidy_text)
y = df.label
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=.2, random_state=12)

tree = DecisionTreeClassifier(max_depth=5)
tree.fit(X_train, y_train)

tree.score(X_train, y_train)
```

```python
X_train, X_test, y_train, y_test = train_test_split(X2, y, test_size=.2, random_state=12)

tree = DecisionTreeClassifier(max_depth=5)
tree.fit(X_train, y_train)

tree.score(X_train, y_train)
```
